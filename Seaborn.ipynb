```python
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data: Ages of a group of people
ages = [22, 25, 27, 22, 35, 34, 40, 24, 33, 31, 29, 38, 30, 24, 26, 22, 27, 28, 30, 29]

# Create the plot
sns.histplot(ages, kde=True, bins=10, color='blue')

# Customizing the plot
plt.title('Age Distribution of a Group')
plt.xlabel('Age')
plt.ylabel('Frequency')

# Display the plot
plt.show()
```
